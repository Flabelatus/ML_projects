## Training and Test Dataset Split

```python
prices = data["PRICE"]
features = data.drop('PRICE', axis=1)

X_train, X_test, y_train, y_test = train_test_split(
    features,
    prices,
#     test size of 20%
    test_size=0.2,
    random_state=10
)

# % of training set
len(X_train) / len(features)
# or
print(X_train.shape[0] / features.shape[0])

# % of test dataset
len(X_test) / len(features)
# or
print(X_test.shape[0] / features.shape[0])
```

## Multivariable Regression

```python
regr = LinearRegression()
regr.fit(X_train, y_train)

# Print r-squared for training and test datasets
print('Training data r-squared:', regr.score(X_train, y_train))
print('Test data r-squared:', regr.score(X_test, y_test))

print("Intercept:", regr.intercept_)
pd.DataFrame(data=regr.coef_, index=X_train.columns, columns=['coef'])

# After printing this features and coefs, you get the values that impact price
# 000s. e.g. RM (number of rooms) increase has coef value of 3.1 meaning that
# the price of the house would be 3.1 x 1000 = 3.100 dollars increased per each
# added room in the house on average.
```
